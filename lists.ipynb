```python
#find the sum of all even numbers in a list
l=[3,4,2,1]
even=0
odd=0
for i in l:
    if  i%2==0:
        even=even+i
    else:
        odd=odd+i
print("sum of even: ",even)
print("sum of odd",odd)
```

```python
#count the vowels in a list
characters=['c','b','a','k','e']
count=0
for char in characters:
    if char.lower() in ['a','e','i','o','u']:
        count+=1
       
print("number of count of vowels",count )
```

```python
#count  the occurrences of each element in a list
l="samarth kadli"
freq={}
for ele in l:
    if ele in freq:
        freq[ele]+=1
    else:
        freq[ele]=1
print("occurrence of a each element in a list",freq)
```
